```python
import rasterio
import numpy as np
from rasterio.plot import show

dataset = rasterio.open('/home/yulya/2015-12-01_layer4')
band4 = dataset.read(1)
print("Канал 4")
show(band4)

dataset = rasterio.open('/home/yulya/2015-12-01_layer8')
band8 = dataset.read(1)
print("Канал 8")
show(band8)

import numpy
import matplotlib.pyplot as plt
%matplotlib inline

numpy.seterr(divide = 'ignore', invalid = 'ignore')
ndvi1 = (band8.astype(float) - band4.astype(float)) / (band8 + band4)
print("Изображение с примененным NDVI")
show(ndvi1)

sz = ndvi1.shape
rows = sz[0]
cols = sz[1]
for i in range(rows):
    for j in range(cols):
        if (ndvi1[i][j] >= -0.75 and ndvi1[i][j] <= -0.15):
            ndvi1[i][j] = -1
        elif (ndvi1[i][j]>= -0.1 and ndvi1[i][j] <= 0.1):
            ndvi1[i][j] = 1
        else:
            ndvi1[i][j] = 0.5
print("Фиолетовый - вода")
print("Желтый - дома, дороги, снег")
print("Зеленый - зеленые массивы")
show(ndvi1)
```
