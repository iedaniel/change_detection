Нью-Йорк, лето. Алгоритм сработал хорошо. Видны мосты через реку и пристани.

```python
import rasterio
import numpy as np
from rasterio.plot import show

dataset = rasterio.open('/home/yulya/layer4')
band4 = dataset.read(1)
print("Канал 4")
show(band4)

dataset = rasterio.open('/home/yulya/layer8')
band8 = dataset.read(1)
print("Канал 8")
show(band8)

import numpy
import matplotlib.pyplot as plt
%matplotlib inline

numpy.seterr(divide = 'ignore', invalid = 'ignore')
ndvi = (band8.astype(float) - band4.astype(float)) / (band8 + band4)
print("Изображение с примененным NDVI")
show(ndvi)

sz = ndvi.shape
rows = sz[0]
cols = sz[1]
for i in range(rows):
    for j in range(cols):
        if (ndvi[i][j] >= -0.75 and ndvi[i][j] <= -0.15):
            ndvi[i][j] = -1
        elif (ndvi[i][j]>= -0.1 and ndvi[i][j] <= 0.1):
            ndvi[i][j] = 1
        else:
            ndvi[i][j] = 0.5
print("Фиолетовый - вода")
print("Желтый - дома, дороги, снег")
print("Зеленый - зеленые массивы")
show(ndvi)
```
